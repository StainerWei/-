```python
import akshare as ak

stock_industry_category_cninfo_df = ak.stock_industry_category_cninfo(symbol="申银万国行业分类标准")
```

```python
industry_2 = stock_industry_category_cninfo_df[stock_industry_category_cninfo_df['分级']==2]
```

```python
industry_2_merge = industry_2[['父类编码','类目名称']]
industry_2_merge.rename(columns={'父类编码':"类目编码",'类目名称':'类目名称_二级'},inplace=True)
```

```python
stock_industry_category_cninfo_df
```

```python
industry_name = pd.merge(stock_industry_category_cninfo_df,industry_2_merge,on=['类目编码'],how='inner')
```

```python
industry_name
```

```python
industry_sec_selection=['医药生物',
                        '商业贸易',
                        '家用电器',
                        '机械设备',
                        '汽车',
                        '电子',
                        '电气设备',
                        '计算机',
                        '通信',
                        '食品饮料']
```

```python
ins = list(industry_name[industry_name['类目名称'].isin(industry_sec_selection)]['类目名称_二级'])
```
